```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# K-Means Clustering

```python
from sklearn import datasets ## Imports dataset from scikit learning
data = datasets.load_iris()
```

```python
print(data.DESCR)
```

```python
data.feature_names
```

```python
#define the ata/ predictiors as the preset feature names
X = pd.DataFrame(data.data, columns = data.feature_names)

#Put the target (Class of Flower) in another dataframe
y = data.target
```

```python
X.head()
```

```python
y[:5]
```

```python
x_ndarray = np.array(X)
x_ndarray[:5]
```

```python
plt.scatter(x_ndarray[:,0],x_ndarray[:,1],c=y,s=50,cmap="viridis")
```

```python
plt.scatter(x_ndarray[:,0], x_ndarray[:,1],c='b',s=50,cmap="viridis")
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(x_ndarray)
y_kmeans = kmeans.predict(x_ndarray)
```

```python
plt.scatter(x_ndarray[:,0],x_ndarray[:,1],c=y_kmeans,s=50,cmap="viridis")

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1], c='black',
           s=200, alpha=0.5)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.scatter(x_ndarray[:,0],x_ndarray[:,1],c=y)
ax2.scatter(x_ndarray[:,0],x_ndarray[:,1],c=y_kmeans)
```
